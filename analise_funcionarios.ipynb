## Objetivo
Realizar a análise exploratória e o tratamento de um conjunto de dados, identificando valores ausentes, duplicados e inconsistências.

```python
import pandas as pd
```

#### Carga de Dados

```python
df = pd.read_excel("./datasets/dataset_tratamento_dados.xlsx")
df.to_csv('dataset_tratamento_dados.csv', index=False)
```

```python
df.info()
```

#### EDA

```python
df.describe()
```

```python
df
```

```python
df = df.drop(columns=['ID'])
```

```python
# Substitui valor nulo de Regiao com moda, pois para variáveis categóricas é a única forma.
df['Regiao'] = df['Regiao'].fillna(df['Regiao'].mode()[0])
```

```python
# Filtra pelo salário negativo
df.loc[df['Salario'] < 0, 'Salario'] = None

# Substitui valor nulo de Salarios com a mediana, pois tem a presença de outliers, sendo assim estatisticamente não ideal fazer com a média;
df['Salario'] = df['Salario'].fillna(df['Salario'].median())
```

```python
# Substitui valor nulo de Avaliacao com mediana, pois como é uma variável quantitativa contínua limitada até 10 é o ideal.
df['Avaliacao'] = df['Avaliacao'].fillna(df['Avaliacao'].median())
```

```python
df.info()
```

```python
# verifindo se tem valores duplicados
df.duplicated(keep=False)
```

```python
# Remove valores duplicados
df = df.drop_duplicates()
```

```python
# valores únicos
df['Regiao'].unique()
```

```python
# Cria um dataset com formato xlsx
df.to_excel('./datasets/dataset_tratado.xlsx', index=False)
```

#### Após a análise exploratória e o tratamento dos dados, foi obtido um conjunto de dados mais consistente e adequado para análises futuras, reduzindo inconsistências e valores inválidos.
